### YoloV11


```python
import torch
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
ds_path = r"F:/JAY/FinalProj/FinalProj-final-.v3i.yolov11/data.yaml"
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

epochs = 5
results = model.train(
    data=ds_path,  
    epochs=epochs,           
    imgsz=640,           
    lr0=0.01,           
    momentum=0.9,       
    optimizer='SGD',          
    weight_decay=0.0005,
    lrf=0.01
)
```

```python
modelVal = model.val()  # evaluate model performance on the validation set
```

```python
from ultralytics import YOLO

Yolov11_Saved = "F:/JAY/FinalProj/Yolov11best.pt"

# Load the saved YOLOv11 model
Yolomodel = YOLO(Yolov11_Saved)
```

```python
import os
img = r"f:\JAY\FinalProj\360_F_136163274_imrNPY1RqqJ1Q0wYRT6bgG2e10Q3hXVG.jpg"
```

```python
import matplotlib.pyplot as plt

# Prediction with display
source_input = img
results = Yolomodel.predict(source=source_input)

# Plotting the results
for result in results:
    # Convert BGR to RGB (OpenCV uses BGR by default)
    plotted_img = result.plot()
    plotted_img = cv2.cvtColor(plotted_img, cv2.COLOR_BGR2RGB)

    # Display with matplotlib
    plt.figure(figsize=(12, 8))
    plt.imshow(plotted_img)
    plt.axis('off')
    plt.show()
```

### RT-DETR

```python
from ultralytics import RTDETR

# RT-DETR-l model
rtdetr_model = RTDETR("rtdetr-l.pt")

# Train 
rtdetr_results = rtdetr_model.train(
    data=ds_path,         
    epochs=10,           
    imgsz=640,            
    batch=2,              
    workers=2,            
    device=0,             
    cache=False,           
    save=True,            
    amp=True              
)
```

```python
from ultralytics import RTDETR

RTDETR_Saved = "F:/JAY/FinalProj/RTDETRbest.pt"

# Load the saved YOLOv11 model
RTDETRmodel = RTDETR(RTDETR_Saved)
```

```python
# Predict the image using the RT-DETR model
rtdetr_results = RTDETRmodel.predict(source=img)

# Plot and display the results
for result in rtdetr_results:
    plotted_img = result.plot()
    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(plotted_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
